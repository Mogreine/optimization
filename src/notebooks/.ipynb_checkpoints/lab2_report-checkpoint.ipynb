## Общее задание

1. Сгенерировать с использованием библиотеки pandas csv файл, содержащий 3 столбца: $x_1$, $x_2$, $y$ (> 400 строк). Функция согласно варианту: $y=0.01\tan(x_1)x_2^{2}$. Количество значений не менее 100 (генерация с использованием функции `linspace`).
2. Открыть файл с использованием библиотеки pandas и построить графики: $y(x_1)$ ($x_2$ - константа), $y(x_2)$ ($x_1$ - константа). Графики необходимо построить с использованием matplotlib (для функции y отобразить точки на графике).
3. Вывести для каждого столбца ($x_1$, $x_2$, $y$): среднее, минимальное и максимальное значения.
4. Сохранить в новый csv файл те строки, для которых выполняется условие: $x_1 < E[x_1]$ или $x_2 < E[x_2]$
5. С использованием mplot3D построить 3D график функции $y(x_1, x_2)$ в отдельном окне.

## Выполнение работы
### Пункт 1
Для начала имортируем нужные для работы библиотеки:

```python
# data analysis and wrangling
import numpy as np
import pandas as pd

# visualization
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

Напишем функцию для вычисления $y$.

```python
def my_func(x1, x2):
    return 0.01 * np.tan(x1) * x2 ** 2
```

Для генерации чисел воспользуемся методом `linspace` из библиотеки `numpy`, который возвращается равномено распределенные числа из заданного интервала. 

```python
x1_arr = np.linspace(0, math.pi / 2, num=500, endpoint=False)
x2_arr = np.linspace(0, 10, num=500)
```

Затем вычислим заданную по варианту функцию и сохраним данные в `DataFrame`, воспользуемся методом `to_csv`, чтобы сохранить данные в формате `csv`.

```python
y_arr = [my_func(x1, x2) for x1, x2 in zip(x1_arr, x2_arr)]
df = pd.DataFrame({
    'y': y_arr,
    'x1': x1_arr,
    'x2': x2_arr
})
df.to_csv('../../data/lab2.csv', index=False)
print('Done!')
```

### Пункт 2
Считаем данные с помощью метода `read_csv` библиотеки `pandas`.

```python
data = pd.read_csv('../../data/lab2.csv')
```

Возьмем $x_1=10$, $x_2=\frac{\pi}{2}$ для отрисовки графиков.

```python
x1_0 = data['x1'][499]
x2_0 = data['x2'][499]
y1_arr = [my_func(x1, x2_0) for x1 in data['x1']]
y2_arr = [my_func(x1_0, x2) for x2 in data['x2']]
```

Опишем необходимые нам методы для работы с графиками из модуля `pyplot`:
- `figure` - создает новый рисунок на котором будут отображаться графики;
- `subplot` - указывает место данного подграфика на общей картинке и сообщает о том, что дальнейшие методы должны применяться именно к этому подграфику.
- `plot` - отрисовывает график, первые 2 параметра это $x$ и $y$ соответственно, 3 - стиль графика, задает способ отображения точек;
- `title` - задает название графику;
- `tight_layout` - предотвращает накладывание графиков друг на друга;
- `show` - показывает получившуюся картинку с графиками.

```python
plt.figure()

plt.subplot(211)
plt.plot(data['x1'], y1_arr, 'ro')
plt.title(f'y=0.01tg(x){x2_0}')

plt.subplot(212)
plt.title('y=0.01tg({:.2f})x'.format(x1_0))
plt.plot(data['x2'], y2_arr, 'bo')

plt.tight_layout()
plt.show()
```

### Пункт 3
Получить минимальное, среднее и максимальное значение столбца в `DataFrame` можно с помощью методов `min`, `mean` и `max` соответственно.

```python
cols = list(data)
col_min = [data[col].min() for col in cols]
col_mean = [data[col].mean() for col in cols]
col_max = [data[col].max() for col in cols]
print('\ty\tx1\tx2')
print('min:\t{:.2f}\t{:.2f}\t{:.2f}'.format(col_min[0], col_min[1], col_min[2]))
print('mean:\t{:.2f}\t{:.2f}\t{:.2f}'.format(col_mean[0], col_mean[1], col_mean[2]))
print('max:\t{:.2f}\t{:.2f}\t{:.2f}'.format(col_max[0], col_max[1], col_max[2]))
```

### Пункт 4
Сначала оставим лишь те строки, которые которые удовлетворяют условию $x_1 < E[x_1]$ или $x_2 < E[x_2]$. А затем сохраним данные с помощью уже описанного метода `to_save`.

```python
df_to_save = data.loc[(data['x1'] < col_mean[1]) | (data['x2'] < col_mean[2])]
df_to_save.to_csv('../../data/lab2_2.csv', index=False)
print('Done!')
```

### Пункт 5
Подготовим данные. Чтобы отрисовать поверхность нужно каждой точке $x_1$ сопоставить все точки $x_2$. Для получения такой матрицы воспользуемся методом `meshgrid`.

```python
x = data['x1'].values
y = data['x2'].values
x, y = np.meshgrid(x, y)
```

Теперь осталось только расчитать значаение функции от каждой из пары значений матрицы.

```python
z = np.array(my_func(np.ravel(x), np.ravel(y)))
z = z.reshape(x.shape)
```

Для отрисовки поверностей придется воспользоваться другими функциями, по сравнению с 2d графиками. Опишем необходимые методы.
- `gca` - указывает тип задания координать (2d/3d/полярные);
- `plot_surface` - аналогично `plot`;
- `set_xlabel` - задает название оси.

```python
ax = plt.figure().gca(projection='3d')
ax.plot_surface(x, y, z)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('График поверхности')

plt.show()
```
