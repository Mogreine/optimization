Произведем изначальную инициализацию

```python
import numpy as np
import cmath
import random as rnd

# visualization
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

cos = np.cos
sin = np.sin
ln = np.log
sqrt = np.sqrt


# unimodal
def f1(x: float):
    """f1"""
    return x * x / 2 + 2 * x - 3, x


# unimodal
def f1_np(x: float):
    """f1"""
    return x ** 2 / 2 + 2 * x - 3, x


# unimodal
def f4(x: float):
    """f4"""
    return -(16 * x ** 2 - 24 * x + 5) * np.e ** (-x), 0

# unimodal
def f4_np(x):
    """f4_np"""
    return -(16 * x ** 2 - 24 * x + 5) * np.exp(-x), 0


def f13(x: float):
    """f13"""
    return -(x ** (2 / 3)) - (1 - x ** 2) ** (1 / 3), 0


def f13_np(x):
    """f13_np"""
    return -np.power(x, 2 / 3) - np.power(1 - x ** 2, 1 / 3), 0


# not unimodal, local minimum
def f5(x: float):
    """f5"""
    return -(1.4 - 3 * x) * sin(18 * x), 0


# not unimodal, global minimum
def f7(x: float):
    """f7"""
    return sin(x) + sin(10 / 3 * x) + ln(x) - 0.84 * x + 3, 0
```

Протестируем метод брента на унимодальных функциях. $f_1(x) = x^2 + 2x - 3$, $f_2(x)=-(16x^2 - 24x + 5)e^{-x}$ и $f_3(x) = -x ^{\frac{2}{3}}-(1 - x^2)^{\frac{1}{3}}$. Их графики:

```python
bound1 = (-6, 2)
step1 = (bound1[1] - bound1[0]) / 50
x1 = np.arange(bound1[0], bound1[1], step1)
y1 = f1_np(x1)[0]

bound2 = (1.9, 3.9)
step2 = (bound2[1] - bound2[0]) / 50
x2 = np.arange(bound2[0], bound2[1], step2)
y2 = f4_np(x2)[0]

bound3 = (0, 1)
step3 = (bound3[1] - bound3[0]) / 50
x3 = np.arange(bound3[0], bound3[1], step3)
y3 = f13_np(x3)[0]

plt.figure(figsize=(8, 16))

plt.subplot(311)
plt.plot(x1, y1)
plt.title('f(x) = x^2 + 2x - 3')
plt.grid(True)

plt.subplot(312)
plt.plot(x2, y2)
plt.title('f(x) = -(16x^2 - 24x + 5)exp(-x)')
plt.grid(True)
plt.yticks(np.arange(-2.4, -4.2, -0.2))

plt.subplot(313)
plt.plot(x3, y3)
plt.title('f(x) = -x^(2 / 3) - (1 - x^2)^(1 / 3)')
plt.grid(True)
plt.yticks(np.arange(-1, -1.6, -0.1))

plt.tight_layout()
plt.show()
```

Перечислим минимумы функций, в которые мы будем пыться сойтись: $x_1^*=-2$, $x_2^*=2.868034$ $x_3^*=0.70711$. Нарисуем графики зависимости $\log|x - x^*|$ от числа итераций. Для поиска минимума используется алгоритма Брента.

```python
def brent(oracle, l: float, r: float, x_true: float, eps: float = 1e-8):
    x1 = l
    x3 = r
    x2 = (l + r) / 2 + 1e-8
    f1 = oracle(x1)[0]
    f2 = oracle(x2)[0]
    f3 = oracle(x3)[0]
    d, d_prev = x3 - x1, x3 - x1
    u = x1
    u_prev = x3
    K = (3 - sqrt(5)) / 2
    err = []
    while abs(u - u_prev) > eps:
        u_prev = u
        d_pr_prev, d_prev = d_prev, d
        u = x2 - ((x2 - x1) ** 2 * (f2 - f3) - (x2 - x3) ** 2 * (f2 - f1)) / \
            (2 * (x2 - x1) * (f2 - f3) - 2 * (x2 - x3) * (f2 - f1) + 1e-8)

        # golden section chooses 'u' if parabola is failing
        if not (x1 < u < x3) or abs(x1 - u) < eps or abs(x3 - u) < eps or abs(u - u_prev) > d_pr_prev / 2:
            if u_prev < (x1 + x3) / 2:
                u = x1 + K * (x3 - x1)
                d = x3 - u_prev
            else:
                u = x3 - K * (x3 - x1)
                d = u_prev - x1
        else:
            d = abs(u - u_prev)
        
        d = abs(u - u_prev)
        
        fu = oracle(u)[0]
        if u > x2:
            if fu > f2:
                x3 = u
                f3 = fu
            else:
                x1 = x2
                f1 = f2
                x2 = u
                f2 = fu
        else:
            if fu > f2:
                x1 = u
                f1 = fu
            else:
                x3 = x2
                f3 = f2
                x2 = u
                f2 = fu
        err.append(abs(u - x_true))
    return u, err

x_true = [-2 + 1e-10, 2.868034 + 1e-10, 1 / sqrt(2) + 1e-10]
log_err1 = np.log10(brent(f1, bound1[0], bound1[1], x_true[0])[1])
iters1 = [*range(1, log_err1.shape[0] + 1)]

log_err2 = np.log10(brent(f4, bound2[0], bound2[1], x_true[1])[1])
iters2 = [*range(1, log_err2.shape[0] + 1)]

log_err3 = np.log10(brent(f13, bound3[0], bound3[1], x_true[2])[1])
iters3 = [*range(1, log_err3.shape[0] + 1)]

plt.figure(figsize=(8, 16))

plt.subplot(311)
plt.plot(iters1, log_err1)
plt.title('Зависимость ошибки от числа итераций')
plt.grid(True)
plt.xticks(iters1)

plt.subplot(312)
plt.plot(iters2, log_err2)
plt.grid(True)
plt.xticks(iters2)

plt.subplot(313)
plt.plot(iters3, log_err3)
plt.grid(True)
plt.xticks(iters3)

plt.tight_layout()
plt.show()
```

Как и ожидалось, на унимодальных функциях метод Брента сходится в глобавльный минимум, причем достаточно быстро. Скачки, которые приводят к понижению точности - это переход метода с метода парабол на золотое сечение.
